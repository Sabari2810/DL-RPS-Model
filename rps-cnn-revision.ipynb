```python
import numpy as np
import pandas as pd
import os
from PIL import Image,ImageOps
import matplotlib.pyplot as plt
```

```python
rock_path = "../input/rock-paper-scissor-images/r_p_s/r/"
paper_path = "../input/rock-paper-scissor-images/r_p_s/p/"
scissor_path = "../input/rock-paper-scissor-images/r_p_s/s/"
```

```python
data = []
labels = []
```

```python
i = 0
for path in [rock_path,paper_path,scissor_path]:
    Images = os.listdir(path)
    for image in Images:
        curr_img = Image.open(path+"/"+image)
        curr_img = ImageOps.grayscale(curr_img)
        curr_img = np.asarray(curr_img)
        data.append(curr_img)
        labels.append(i)
    i+=1
```

```python
X = np.array(data)
y = np.array(labels)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.1)
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
from sklearn.preprocessing import LabelBinarizer
```

```python
encoder = LabelBinarizer()
```

```python
y_train = encoder.fit_transform(y_train)
```

```python
y_train
```

```python
y_test = encoder.fit_transform(y_test)
```

```python
X_train.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,Dense,MaxPool2D,Flatten
```

```python
model = Sequential()
model.add(Conv2D(filters=32,kernel_size=(3,3),activation="relu",input_shape=(380,400,1)))
model.add(MaxPool2D(pool_size=(4,4)))

model.add(Conv2D(filters=32,kernel_size=(3,3),activation="relu",input_shape=(380,400,1)))
model.add(MaxPool2D(pool_size=(4,4)))

model.add(Flatten())

model.add(Dense(128,activation="relu"))
model.add(Dense(3,activation="softmax"))
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy")
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor="val_loss",verbose=1,patience=2)
```

```python
model.fit(X_train,y_train,epochs=10,validation_data=(X_test,y_test),callbacks=[early_stop])
```
